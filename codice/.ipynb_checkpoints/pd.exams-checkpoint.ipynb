```python
## Question
# How many students are enrolled in the class?

numStudentsEnrolled = len(dfs)
print(numStudentsEnrolled)
```

```python
## Question
# How many students have taken the exam at least once?

numStudentsTakenExam = len(dfr['CP'].unique())
print(numStudentsTakenExam)
```

```python
## Question
# How many students did not take any exam?
numStudentsNoExams = numStudentsEnrolled - numStudentsTakenExam
print(numStudentsNoExams)
```

```python
# In several questions we need to have the information on the grades per student per exam
# I prepare the dataframe that has this information

# 1. add the information on the exam to the result dfr
# 2. compute the grade for each student / exam

# combine dfr and dfe to add the info about the exam each question belongs to > merge
dfre = dfr.merge(dfe,left_on='question', right_on='QID').drop(['QID'], axis=1)
# for each student exam pair, compute the grade by adding the values of the questions > groupby
# resetting the index to get a flat table
# dfres = dfre.groupby(['CP','EID']).agg({'value': "sum"}).reset_index()
# renaming the column for clarity
# dfres.rename(columns={'value':'grade'}, inplace=True)
# or equivalent and more compact
# 
dfres = dfre.groupby(['CP','EID'])['value'].sum().reset_index(name='grade')

print(dfres)
```

```python
## Question
# For the entire class (all exams), what is the highest grade, the lowest (passing) grade, the average (2 decimal digits) and median grades?

passing = dfres[dfres['grade'] > 17]['grade']

# Solution 1: use describe and extra
stats = passing.describe()
avgGradeYear = round(stats['mean'], 2)
minGradeYear = stats['min']
maxGradeYear = stats['max']
medGradeYear = stats['50%']

print(maxGradeYear, minGradeYear, avgGradeYear, medGradeYear)

# Solution 2: compute
# I want to have the values > min, max, median, mean
minGradeYear = passing.min()
maxGradeYear = passing.max()
avgGradeYear = round(passing.mean(), 2)
medGradeYear = passing.median()

print(maxGradeYear, minGradeYear, avgGradeYear, medGradeYear)

# Solution 3:
stats = passing.agg(['min', 'max', 'mean', 'median'])
avgGradeYear = round(stats['mean'], 2)
minGradeYear = stats['min']
maxGradeYear = stats['max']
medGradeYear = stats['median']
print(maxGradeYear, minGradeYear, avgGradeYear, medGradeYear)
```

```python
## Question
# For each exam, what is the highest grade, the lowest (passing) grade, the average (2 decimal digits) and median grades?

# it is necessary to get the information BY exam > groupby extracting the statistical elements we need
dfres[dfres['grade'] > 17][['EID','grade']].groupby(['EID']).agg(['min', 'max', 'mean', 'median'])
# or 
result = dfres[dfres['grade'] > 17].groupby('EID')['grade'].agg(
    minGrade='min',
    maxGrade='max',
    avgGrade=lambda x: round(x.mean(), 2),
    medGrade='median'
)
result
```

```python
# create a tabular view of the exams for each student, including those who have never attended an exam 

# reshape the table > pivot
dfse = dfres.pivot(index='CP', columns='EID', values='grade').reset_index()
# add the students with no exam
dfse_all = pd.merge(dfs, dfse, how="left")
dfse_all
```

```python
## Add a column with the final grade if they passed

# exam names
examlist = dfe['EID'].unique()
examlist_rev = list(reversed(examlist)) #first the most recent exam

def getLatestPassingGrade(row):
    # for every exam outcome
    for exam in examlist_rev:
        # if there is a value (it is not a NaN) and is sufficient
        if pd.notna(row[exam]) and row[exam] >= 18:
            # that is the grade we are looking for
            return row[exam]
    # no valid grade
    return None

# for all rows, apply the function  getLatestPassingGrade
dfse_all['finalGrade'] = dfse_all.apply(getLatestPassingGrade, axis=1)
dfse_all
```

```python
## Question
# At the end of the year, looking at the final grade, what is the highest grade, 
# the lowest (passing) grade, the average  (2 decimal digits) and median grades?

finalResult = dfse_all['finalGrade'].agg(
    minGradeYear='min',
    maxGradeYear='max',
    avgGradeYear=lambda x: round(x.mean(), 2),
    medGradeYear='median')
print(finalResult['minGradeYear'],finalResult['maxGradeYear'],finalResult['avgGradeYear'],finalResult['medGradeYear'])
```

```python
## Question
# How many students did not take any exam?
dfnoexams = dfse_all[dfse_all[['E1', 'E2', 'E3', 'E4', 'E5']].isna().all(axis=1)]
print(len(dfnoexams))
```

```python
## Question
# Overall, how many students have passed the exam?

numPassedStudents = dfse_all['final_grade'].notna().sum()
print(numPassedStudents)
```

```python
## Question
# Display the students who passed the exam (CP) and their grade

dfse_all[dfse_all['finalGrade']>=18][['CP','finalGrade']]
```

```python
## Question
# What are the ratio of students passing with respect to enrolled students and with respect to student taking the exam?

ratioPassedEnrolled = round(numPassedStudents*100 / numStudentsEnrolled, 2)
ratioPassedExisting = round(numPassedStudents*100 / numStudentsTakenExam, 2)
print(ratioPassedEnrolled, ratioPassedExisting)
```

```python
## Question
# On average, how many exams did each student take to pass the exam with the _desired_ grade?

def countExamsPerStudent(row):
    # for every exam outcome
    num = 0
    for exam in examlist:
        if pd.notna(row[exam]):
            num += 1
    # no valid grade
    return num
# add a new column   
dfse_all['numExams'] = dfse_all.apply(countExamsPerStudent, axis=1)

avgNumExamsPerStudent = dfse_all['numExams'].mean()
avgNumExamsPerStudent
```

```python
## Question
# For the entire class, how many students repeated the exam at least once?

print(len(dfse_all[dfse_all['numExams'] > 1]))
```
